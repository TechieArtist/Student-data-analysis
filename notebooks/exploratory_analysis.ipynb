```python
import pandas as pd

# Load the dataset
df = pd.read_csv('E:\student_data_analysis\data\processed\processed_student_scores_general.csv')
```

```python
# Display the first few rows of the dataset
df.head()

# Display basic information about the dataset
df.info()

# Statistical summary of the dataset
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Distribution of Math Scores
sns.histplot(df['math_score'], kde=True)
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot for Math Scores
sns.boxplot(x=df['math_score'])
plt.title('Box Plot of Math Scores')
plt.show()
```

```python
# Scatter plot between Weekly Self Study Hours and Math Score
sns.scatterplot(x='weekly_self_study_hours', y='math_score', data=df)
plt.title('Weekly Self Study Hours vs. Math Score')
plt.xlabel('Weekly Self Study Hours')
plt.ylabel('Math Score')
plt.show()
```
